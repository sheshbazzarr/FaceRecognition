```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
```

```python
#   Load numpy arra
data = np.load('./data/data_pca_50_target.npz')
data.files
```

```python
data.allow_pickle = True
```

```python
X = data['arr_0'] # oca data with 50 components 
y = data['arr_1'] # target or dependet variable  
```

```python
X.shape
```

```python
X
```

```python
y
```

## Split the data into train and test

```python
# # split the data into train and test
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify = y)
print(x_train.shape,x_test.shape,y_test.shape,y_train.shape)
```

### Training Machine Learning Model

```python
model_svc = SVC(probability=True)
param_grid = {'C':[0.5,1,10,20,30,50],'kernel':['rbf','poly'],'gamma':[0.1,0.05,0.01,0.001,-.002,0.005],'coef0':[0,1]}
```

```python
model_grid = GridSearchCV(model_svc,param_grid=param_grid,scoring='accuracy',cv=3,verbose=2)
```

```python
model_grid.fit(x_train,y_train)
```

```python
model_grid.best_params_
```

```python
model_final = model_grid.best_estimator_
```

```python
model_final.get_params()
```

## Model Evaluation

```python
y_pred= model_final.predict(x_test) # predicted values
```

```python
y_pred
```

```python
# classifcation report 
cr = metrics.classification_report(y_test,y_pred,output_dict=True)
pd.DataFrame(cr).T
```

```python
# kappa score : my model is good model 
metrics.cohen_kappa_score(y_test,y_pred)
```

```python
# Area under curve (AUC) metrics 
# good model . excellint model 
metrics.roc_auc_score(np.where(y_test=="male",1,0),
                      np.where(y_pred=="male",1,0))
```
