```python
import numpy as np
import cv2 
from glob import glob 
import matplotlib.pyplot as plt
```

```python
fpath = glob('./data/female/*.jpg')
mpath = glob('./data/male/*.jpg')
```

```python
# Step -1 Read Image and Convert to RGB
img = cv2.imread(fpath[0])  # read image BRG
img_rgb =cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #tHIS STEP WILL CONVERT IMAGE FROM brg TO rgb


# Step -2 apply Haar Cascade Classifier
haar =cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
faces_list = haar.detectMultiScale(gray, 1.5, 5)
print(faces_list)

for x,y,w,h in faces_list:
    cv2.rectangle(img_rgb, (x,y), (x+w, y+h), (255,0,0), 2)

    # step -3 crop face
    ro = img_rgb[y:y+h, x:x+w]
    plt.imshow(ro)
    plt.axis('off')
    plt.show()
```

```python
import cv2
import os

# Initialize Haar Cascade
haar = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Ensure output directory exists
output_dir = './crop_data/females/'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Process images
for i in range(len(fpath)):
    try:
        # Step 1: Read image and convert to RGB
        img = cv2.imread(fpath[i])
        if img is None:
            print(f'Unable to read the image at {fpath[i]}')
            continue
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Step 2: Apply Haar Cascade Classifier
        gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
        faces_list = haar.detectMultiScale(gray, 1.5, 5)

        # Step 3: Crop and save detected faces
        for idx, (x, y, w, h) in enumerate(faces_list):
            ro = img[y:y+h, x:x+w]
            
            # Format the filename with zero-padded numbers
            formatted_index = f"{i:06d}"  # Zero-padded 6-digit index
            save_path = os.path.join(output_dir, f'female_{formatted_index}.jpg')
            
            cv2.imwrite(save_path, ro)
            print(f'Image successfully processed: {save_path}')
    except Exception as e:
        print(f'Unable to process the image {fpath[i]}: {e}')

```
